```python
data_state['avg crime'] = np.sum(data_state.iloc[:,2:].values, axis=1)
```

```python
def plot_state(data, state):
    '''
    Function to plot the 
    '''
    f = plt.figure()
    plt.title('Crime-Head-'+state+'-(2001-2011)', color='black')

    data[data['STATE/UT']==state]\
        .iloc[:-1, 1:14]\
        .set_index('CRIME HEAD').T\
        .plot(kind='bar', figsize=(26, 17), ax=f.gca())

    plt.legend(loc='best')
    plt.show()
```

```python
plot_state(data_ut, UTs[0])
```

```python
plot_state(data_ut, UTs[3])
```

```python
plot_state(data_ut, UTs[6])
```

```python
plot_state(data_state, STATES[4])
```

```python
plot_state(data_state, STATES[7])
```

```python
plot_state(data_state, STATES[21])
```

```python
plot_state(data_state, STATES[19])
```
